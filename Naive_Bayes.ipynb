```python
import random
random.seed(42)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('com410-semana3-Iris.csv', header=(0))
data = data.dropna(axis='rows') # Remover dados inválidos
```

```python
classes = np.array(pd.unique(data[data.columns[-1]]), dtype=str) #Armazena as especies na classe
print(classes) #Certificar que dados foram armazenados no array
```

```python
atributos = list(data.columns) # Adicionar as colunas de atributos das flores a variável "atributos"
data.head(10) #Exibir os primeiros 10 elementos da matriz
```

```python
data= data.to_numpy() #Converte o Data Frame em um array
nrow, ncol = data.shape # Armazena a quantidade de linhas e colunas nas variáveis nrow e ncol;
```

```python
y = data[:,-1] #Amazenando na variável "y" os valores do atributo de especies
x= data[:,0:ncol-1] #Armazenando na variável "x" os valores dos atributos de largura e comprimento das sépalas e pétalas
```

##Selecionando o conjunto de dados para treino e teste

```python
from sklearn.model_selection import train_test_split
p= 0.7 #Separando a fração para o conjunto de treinamento
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=p, random_state=42)
```

###Aplicando o processo Gaussiano e verificando acurácia do modelo.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix

model = GaussianNB()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
score = accuracy_score(y_pred, y_test)
matriz_confusão = confusion_matrix(y_test,y_pred)
print('Acurácia: ', score)
matriz_confusão
```

###Mostrando Data Frame com valores reais e os que foram predictos pelo modelo.

```python
df = pd.DataFrame({'Valores Reais':y_test, 'Valores Predictos':y_pred})
df
```
